This is a calculator used for calculating Financial Freedom by using the concept of Perpetuity. 

How much money do you need every month to survive?

```python
money_needed_every_month= 200000
```

Money needed in a year to survive



```python
money_needed_every_year = money_needed_every_month*12
money_needed_every_year
```

We will take the bare minimum interest rates offered. You can change this accordingly. Please do not use 6% but instead write 0.06 

```python
fd_interest_rate= 0.06
```

Now we will calculate the total capital that we have to invest today to keep the show running for the rest of your life!!!
This below is the formula for perpetuity. For more information please visit https://corporatefinanceinstitute.com/resources/knowledge/finance/perpetuity/

```python
total_capital = money_needed_every_year/fd_interest_rate
total_capital
```

```python
#mutual_funds_interest_rate= 0.12
```

```python
for i in range(100):
  total_capital = total_capital* (1 + fd_interest_rate)
  print("\nYear " , i ," \nTotal Capital after Interest = \t", total_capital )
  total_capital = total_capital - money_needed_every_year
  print("Total Capital after Withdrawal = ", total_capital )
```
